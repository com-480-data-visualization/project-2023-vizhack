```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cig_pc_pd = pd.read_csv('IHME_GBD_2019_SMOKING_TOB_1990_2019_CIG_PC_Y2021M05D27.CSV')
num_smokers_pd = pd.read_csv('IHME_GBD_2019_SMOKING_TOB_1990_2019_NUM_SMOKERS_Y2021M05D27.CSV')
pred_pd = pd.read_csv('IHME_GBD_2019_SMOKING_TOB_1990_2019_PREV_Y2021M05D27.CSV')
```

```python
num_smokers_pd.head(5)
```

The graph shows the trend of total number of smokers from 1990 to 2019

```python
years = num_smokers_pd.groupby(by=['year_id']).sum()

fig = plt.figure(figsize=(16,4))
ax = sns.barplot(x=years.index,y=years.val, width=0.8)
```

The plot shows the trend of total number of male smokers from 1990 to 2019

```python
male_years = num_smokers_pd[num_smokers_pd['sex_name']=='Male']
myears = male_years.groupby(by='year_id').sum()
fig = plt.figure(figsize=(16,4))
ax = sns.barplot(x=myears.index,y=myears.val, width=0.8)
```

```python
fig = plt.figure(figsize=(16,4))
ax = sns.lineplot(x=myears.index,y=myears.val)
```

```python
female_years = num_smokers_pd[num_smokers_pd['sex_name']=='Female']
fmyears = female_years.groupby(by='year_id').sum()
fig = plt.figure(figsize=(16,4))
ax = sns.barplot(x=fmyears.index,y=fmyears.val, width=0.8)
```

```python
fig = plt.figure(figsize=(16,4))
ax = sns.lineplot(x=fmyears.index,y=fmyears.val)
```

```python
y1990 = num_smokers_pd[num_smokers_pd['year_id']==1990]
y1990_ = y1990.groupby(by='location_name').sum()
y1990_ = y1990_.sort_values(by=['val'],ascending=False)
fig = plt.figure(figsize=(16,4))
ax = sns.barplot(x=y1990_.head(7).index,y=y1990_.head(7).val, width=0.8)
```

```python
y1990 = num_smokers_pd[num_smokers_pd['year_id']==2019]
y1990_ = y1990.groupby(by='location_name').sum()
y1990_ = y1990_.sort_values(by=['val'],ascending=False)
fig = plt.figure(figsize=(16,4))
ax = sns.barplot(x=y1990_.head(7).index,y=y1990_.head(7).val, width=0.8)
```
