```python
import pandas as pd
```

```python
df = pd.read_csv('../data/initial_data/SMOKING_TOB_1990_2019_PREV.CSV')
df.head()
```

```python
df_smoking = df[['location_name', 'val', 'year_id', 'sex_name', 'age_group_name']]
df_smoking.head()
```

```python
df_smoking = df_smoking[df_smoking['age_group_name'] == '15+ years']
df_smoking['val'] = df_smoking['val']*100
df_smoking = df_smoking[df_smoking['sex_name'] == "Female"]
df_subset = df_smoking[['location_name', 'val', 'year_id']]
path = '../data/multiline_data/multiline_chart_Female.CSV'
df_subset.to_csv(path, index=False)
```
